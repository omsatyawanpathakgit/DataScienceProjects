````python
# remove duplicates
df.drop_duplicates(inplace=True)
````

````python


df.columns = df.columns.str.lower()

df.head()




# RAM conversion

df["ram_gb"] = (
    df["ram_gb"]
    .str.replace(" GB","")
    .astype(int)
)



# SSD conversion

df["ssd"] = (
    df["ssd"]
    .str.replace(" GB","")
    .astype(int)
)



# HDD conversion

df["hdd"] = (
    df["hdd"]
    .str.replace(" GB","")
    .astype(int)
)



# Graphic card
df["graphic_card_gb"] = (
    df["graphic_card_gb"]
    .str.extract("(\d+)")
    .fillna(0)
    .astype(int)
)



# Rating conversion

df["rating"] = (
    df["rating"]
    .str.replace(" stars","", regex=False)
    .str.replace(" star","", regex=False)
    .astype(int)
)



# OS bit

df["os_bit"] = (
    df["os_bit"]
    .str.replace("-bit","")
    .astype(int)
)



df.head()



# ==========================================
# 7. CREATE NEW FEATURES
# ==========================================


# Total storage

df["total_storage"] = (
    df["ssd"] + df["hdd"]
)



# Storage type

df["has_ssd"] = np.where(
    df["ssd"]>0,1,0
)


# GPU availability

df["has_gpu"] = np.where(
    df["graphic_card_gb"]>0,1,0
)



# Laptop performance score

df["performance_score"] = (
    df["ram_gb"]*2 +
    df["processor_name"].map({
        "Core i3":3,
        "Core i5":5,
        "Core i7":7,
        "Core i9":9,
        "Ryzen 3":3,
        "Ryzen 5":5,
        "Ryzen 7":7,
        "M1":8,
        "Celeron Dual":2,
        "Pentium Quad":2
    }).fillna(0)
    +
    df["graphic_card_gb"]
)



# customer popularity

df["popularity_score"] = (
    df["number of ratings"] +
    df["number of reviews"]
)


# Total Storage:
df["total_storage"] = (
    df["ssd"] + df["hdd"]
)

#GPU availability:
df["has_gpu"] = np.where(
    df["graphic_card_gb"]>0,
    1,
    0
)



df.drop(
    [
    "processor_name"
    ],
    axis=1,
    inplace=True
)






````

**Problem statement:** Is the average price of ASUS laptops significantly different from the average price of Dell laptops? (Two-sample ztest)

````python
price_asus = df[df["brand"]=="ASUS"]["price"]
price_dell = df[df["brand"]=="DELL"]["price"]



z_stat, p_value = ztest(price_asus, price_dell)

alpha = 0.05

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject H0: Significant difference in average price between ASUS and DELL.")
else:
    print("Fail to reject H0: No significant difference in average price between ASUS and DELL.")
````

**Problem statement**: Is the proportion of laptops that come with an SSD (Solid State Drive) greater than 80%? (one-sample ztest)

````python
from statsmodels.stats.proportion import proportions_ztest



count = df["has_ssd"].sum()
nobs = len(df)

# Hypothesized proportion (80% = 0.80)
hypothesized_prop = 0.80

# One-sample z-test for proportion (alternative="larger" for >)
z_stat, p_value = proportions_ztest(
    count=count,
    nobs=nobs,
    value=hypothesized_prop,
    alternative="larger"   # one-tailed: testing if greater than 80%
)

alpha = 0.05


if p_value < alpha:
    print("Reject H0: The proportion of laptops with SSD is significantly greater than 80%.")
else:
    print("Fail to reject H0: No evidence that the proportion of laptops with SSD exceeds 80%.")
````

**Problem Statement:** Is the proportion of laptops running Windows greater than 85%? (One-sample z-test for proportion)

````python
from statsmodels.stats.proportion import proportions_ztest


windows_count = (df["os"] == "Windows").sum()
total_laptops = len(df)


hyp_prop = 0.85

z_stat, p_value = proportions_ztest(
    count=windows_count,
    nobs=total_laptops,
    value=hyp_prop,
    alternative="larger"
)

alpha = 0.05


if p_value < alpha:
    print("Reject H0: Proportion of Windows laptops is significantly greater than 85%.")
else:
    print("Fail to reject H0: Proportion of Windows laptops does not exceed 85%.")
````

**Problem statement:** Is the proportion of laptops with a touchscreen different between HP and Lenovo? (Two-sample z-test for proportions)

````python

hp_touch = df[df["brand"] == "HP"]["Touchscreen"]
lenovo_touch = df[df["brand"] == "Lenovo"]["Touchscreen"]


count_hp = (hp_touch == "Yes").sum()
count_lenovo = (lenovo_touch == "Yes").sum()


n_hp = len(hp_touch)
n_lenovo = len(lenovo_touch)


z_stat, p_value = proportions_ztest(
    count=[count_hp, count_lenovo],
    nobs=[n_hp, n_lenovo],
    alternative="two-sided"
)

alpha = 0.05


if p_value < alpha:
    print("Reject H0: Significant difference in touchscreen proportion between HP and Lenovo.")
else:
    print("Fail to reject H0: No significant difference in touchscreen proportion.")
````

**Problem statement** : "A laptop company wants to segment its laptops into different customer segments based on specifications, pricing, and customer feedback. Build a clustering model that groups laptops into meaningful categories such as Budget laptops, Professional laptops, Gaming laptops, and Premium laptops."

````python
df.shape
````

````python
cluster_df = df[
[
    "price",
    "ram_gb",
    "graphic_card_gb",
    "total_storage",
    "performance_score",
    "rating",
    "number of ratings",
    "number of reviews",
    "brand",
    "processor_brand",
    "os"
]
]


cat_cols = cluster_df.select_dtypes(include="object").columns
num_cols = cluster_df.select_dtypes(exclude="object").columns


preprocessor = ColumnTransformer(
    transformers=[
        (
            "categorical",
            OneHotEncoder(handle_unknown="ignore"),
            cat_cols
        ),
        (
            "numeric",
            StandardScaler(),
            num_cols
        )
    ]
)

X = preprocessor.fit_transform(cluster_df)

````

````python
from sklearn.cluster import KMeans

wcss = []

for k in range(1,11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X)

    wcss.append(kmeans.inertia_)


plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    wcss,
    marker="o"
)

plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")

plt.show()
````

````python
from sklearn.metrics import silhouette_score

for k in range(2,11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(X)

    score = silhouette_score(X, labels)

    print(f"K = {k} --> Silhouette Score = {score:.4f}")
````

````python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(X)


cluster_df["cluster"] = clusters
````

````python
pd.crosstab(
    cluster_df["cluster"],
    cluster_df["brand"]
)
````

````python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

plt.figure(figsize=(10,6))



cluster_names = {
    0: "Premium / High-Performance Laptops",
    1: "Professional Laptops",
    2: "Budget & Mass-Market Laptops"
}

cluster_labels = pd.Series(clusters).map(cluster_names)


sns.scatterplot(
    x=X_pca[:,0],
    y=X_pca[:,1],
    hue=cluster_labels,
    palette="Set1"
)




plt.xlabel("PC1")
plt.ylabel("PC2")


plt.legend(title="Laptop Segment")
plt.title("Laptop Segmentation using KMeans")
plt.show()
````

**Although the business initially expected four segments (Budget, Professional, Gaming, Premium), Silhouette Analysis indicated that the optimal number of clusters is 3 (score = 0.3278). Therefore, K-Means with K=3 was selected to obtain the most meaningful and well-separated customer segments.**
